```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
from datetime import timedelta
from selenium import webdriver
from urllib.request import urlopen
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
path = '/Users/yichuanhong/Documents/Python Project/Vietnam Avian Flu Outbreak Research/NLP/chromedriver'
url = 'https://www.doisongphapluat.com/search?q=c%C3%BAm+gia+c%E1%BA%A7m'
root = 'https://www.doisongphapluat.com'
df = pd.DataFrame(columns = ['Title', 'URL', 'Date', 'Content'])
driver = webdriver.Chrome(path)
driver.get(url)
```

```python
for i in range(2, 12):
    soup = BeautifulSoup(driver.page_source)
    news = soup.find_all('div', class_ = 'gsc-webResult gsc-result')
    for item in news:
        title = item.find('div', class_ = 'gs-title').text
        url = item.find('a', class_ = 'gs-title')['href']
        response = requests.get(url).text
        soup = BeautifulSoup(response, 'lxml')
        date = soup.find('div', class_ = 'datetime mb-3')
        if date:
            date = datetime.strptime(date.text.split(', ')[1].split(' ')[0], '%d/%m/%Y')
        content = soup.find('article')
        if content:
            content = content.text
        new_row = {'Title': title, 'URL': url, 'Date': date, 'Content': content}
        df = df.append(new_row, ignore_index = True)
    if i < 11:
        next_page = driver.find_element_by_xpath('//*[@id="___gcse_0"]/div/div/div/div[5]/div[2]/div/div/div[2]/div/div[' + str(i) + ']')
        next_page.click()
        time.sleep(2)
```
