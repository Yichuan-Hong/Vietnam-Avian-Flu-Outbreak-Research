```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
from datetime import timedelta
from selenium import webdriver
from urllib.request import urlopen
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
path = '/Users/yichuanhong/Documents/Python Project/Vietnam Avian Flu Outbreak Research/NLP/chromedriver'
url = 'https://baohatinh.vn/search/?q=c%C3%BAm+gia+c%E1%BA%A7m&ch=0&p=1'
root = 'https://baohatinh.vn'
df = pd.DataFrame(columns = ['Title', 'URL', 'Date', 'Content'])
#driver = webdriver.Chrome(path)
#driver.get(url)
```

```python
for i in range(2, 15):
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'lxml')
    news = soup.find_all('li', class_ = 'hed li_search clearfix')
    for item in news:
        article = item.find('div', class_ = 'title_li_politic')
        title = article.find('a')['title']
        sub_url = article.find('a')['href']
        if sub_url.startswith('https'):
            url = sub_url
        else:
            url = root + sub_url
        response = requests.get(url).text
        new_soup = BeautifulSoup(response, 'lxml')
        date = new_soup.find('div', class_ = 'date_news')
        if date:
            date = datetime.strptime(date.text.split(' ')[-1], '%d/%m/%Y')
        content = new_soup.find('div', class_ = 'col615 fl')
        if content:
            content = content.text
        new_row = {'Title': title, 'URL': url, 'Date': date, 'Content': content}
        df = df.append(new_row, ignore_index = True)
    url = 'https://baohatinh.vn/search/?q=c%C3%BAm+gia+c%E1%BA%A7m&ch=0&p=' + str(i)
```
